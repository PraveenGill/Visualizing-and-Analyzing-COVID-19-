# **4. Getting Insights and Visualization**
Now, our dataset is ready to be analyzed, so we would try to find out answer to following questions from COVID-19 dataset:


1.   What is total number of active cases in India?
2.   What is recovery rate and mortality rate?
3. What is total number of confirmed, cured and deaths cases in each state?
4. Which states have highest number of confirmed cases, deaths?
5. What is growth trend of virus?



```python
#active covid cases
covid['ActiveCases']=covid['Confirmed']-(covid['Cured']+covid['Deaths'])
```

```python
covid.tail()
```

```python
#summing death and cured for each state and union teritory using pivot_table()

statewise=pd.pivot_table(covid,values = ['Confirmed','Deaths','Cured'], index = 'State/UnionTerritory', aggfunc= max)
```

```python
#finding recovery rate

statewise['Recovery_Rate']=statewise['Cured']*100/statewise['Confirmed']
```

```python
#finding mortality rate

statewise['Mortality_Rate']=statewise['Deaths']*100/statewise['Confirmed']
```

```python
#sorting data in descending order by Confirmed cases

statewise = statewise.sort_values(by = 'Confirmed', ascending = False)
```

```python
#ploting the statewise table using cmap

statewise.style.background_gradient(cmap = "cubehelix")
```

```python
# top 10 stateswith active cases

top10_activecases=covid.groupby(by = 'State/UnionTerritory').max()[['ActiveCases', 'Date']].sort_values(by = ['ActiveCases'], ascending = False).reset_index()
```

```python
fig= plt.figure(figsize= (16,9))
plt.title("Top 10 states with most active cases in India")
ax= sns.barplot(data= top10_activecases[:10], y='ActiveCases', x='State/UnionTerritory', linewidth=2, edgecolor='black')
```

```python
# top 10 stateswith active cases

top10_acrivecases=covid.groupby(by = 'State/UnionTerritory').max()[['ActiveCases', 'Date']].sort_values(by = ['ActiveCases'], ascending = False).reset_index()

fig= plt.figure(figsize= (16,9))
plt.title("Top 10 states with most active cases in India")
ax= sns.barplot(data= top10_activecases[:10], y='ActiveCases', x='State/UnionTerritory', linewidth=2, edgecolor='black')
plt.xlabel("State/Union Territory")
plt.ylabel("Total no. of active cases")
plt.show()
```

```python
# top 10 staes with maximum deaths

top10_deaths= covid.groupby(by= 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'], ascending=False).reset_index()

fig=plt.figure(figsize=(16,9))
plt.title("Top 10 states with most covid deaths in India")
ax= sns.barplot(data= top10_deaths[:10], y='Deaths', x='State/UnionTerritory', linewidth=2, edgecolor='black')
plt.xlabel('State/Union Territory')
plt.ylabel("Total no. of deaths ")
plt.show()
```

```python
# we discripency in datset so we will plot 12 states to get top 10 statesmaharastra and karnatka have 2 difrent spellings
# that should be resovled during data cleaning so remove it

top10_deaths= covid.groupby(by= 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'], ascending=False).reset_index()

fig=plt.figure(figsize=(16,9))
plt.title("Top 10 states with most covid deaths in India")
ax= sns.barplot(data= top10_deaths[:12], y='Deaths', x='State/UnionTerritory', linewidth=2, edgecolor='black')
plt.xlabel('State/Union Territory')
plt.ylabel("Total no. of deaths ")
plt.show()
```

```python
#growth trend in top 5 states of covid

fig=plt.figure(figsize=(15,9))
ax= sns.lineplot(data=covid[covid['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerela','Tamil Nadu', 'Uttar Pradesh'])],
                            x= 'Date', y='ActiveCases', hue= 'State/UnionTerritory')
ax.set_title('Top 5 affectedstates in India', size=16)
```

Using vaccineation dataset, we would try to find out answers to following questions:


1.   What is the ratio of male and female vaccination?
2.   Which states have maximum number of vaccination?
3. which are least vaccinated states?


```python
# male v/s female vaccination

male = vaccination['Male(Individuals Vaccinated)'].sum()
female = vaccination['Female(Individuals Vaccinated)'].sum()
px.pie(names=['Male','Female'],values=[male,female],title='Male and Female Vaccination')
```

```python
# remove rows with state as India

vaccine1 = vaccine[vaccine.State!='India']
vaccine1
```

```python
vaccine1.rename(columns={"Total Individuals Vaccinated": "Total"}, inplace=True)
vaccine1.head()
```

```python
# most vaccinated states
maxvaccine= vaccine1.groupby('State')['Total'].sum().to_frame('Total')
maxvaccine= maxvaccine.sort_values('Total', ascending=False)[:5]
maxvaccine
```

```python
fig= plt.figure(figsize=(10,5))
plt.title('Top 5 Vaccinated States in India',size=20)
x=sns.barplot(data=maxvaccine.iloc[:10],y=maxvaccine.Total,x=maxvaccine.index,linewidth=2,edgecolor='black')
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```

```python
# least vaccinated states in india
minvaccine=vaccine1.groupby('State')['Total'].sum().to_frame('Total')
minvaccine=minvaccine.sort_values('Total')[:5]
minvaccine
```

```python
fig= plt.figure(figsize=(10,5))
plt.title('Least 5 Vaccinated States in India',size=20)
x=sns.barplot(data=minvaccine.iloc[:10],y=minvaccine.Total,x=minvaccine.index,linewidth=2,edgecolor='black')
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```
